```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

# FOLD 1

```python
setFold1Training = pd.read_csv('act2.1_diabetes_2000_rand_training80_rand_fold1_test.csv')
setFold1Test = pd.read_csv('act2.1_diabetes_2000_rand_training80_rand_fold1_training.csv')

atributos_Fold1Training = setFold1Training[setFold1Training.columns[:-1]]
classes_Fold1Training = setFold1Training[setFold1Training.columns[-1]]

atributos_Fold1Test = setFold1Test[setFold1Test.columns[:-1]]
classes_Fold1Test = setFold1Test[setFold1Test.columns[-1]]

trainingAccuracyValues_Fold1 = []
testAccuracyValues_Fold1 = []
trainingErrorValues_Fold1 = []
testErrorValues_Fold1 = []

n_epoch = range(10, 950, 10)

# ---------- PARA QUE NO IMPRIMA LOS WARNINGS ---------- #
import warnings
warnings.filterwarnings("ignore")
# ------------------------------------------------------ #

for epoch in n_epoch:
    # Build the model
    # clasificador = MLPClassifier(solver='sgd', 
    clasificador = MLPClassifier(solver='lbfgs', 
                    hidden_layer_sizes=(10), 
                    # momentum=0.4,
                    # learning_rate_init=0.001,
                    random_state=42,
                    max_iter = epoch)
    clasificador.fit(atributos_Fold1Training, classes_Fold1Training)
    # record training set accuracy and error
    trainingAccuracyValues_Fold1.append(clasificador.score(atributos_Fold1Training, classes_Fold1Training))
    trainingErrorValues_Fold1.append(1.0 - clasificador.score(atributos_Fold1Training, classes_Fold1Training))
    # record generalization accuracy and error
    testAccuracyValues_Fold1.append(clasificador.score(atributos_Fold1Test, classes_Fold1Test))
    testErrorValues_Fold1.append(1.0 - clasificador.score(atributos_Fold1Test, classes_Fold1Test))

plt.title("Error: FOLD 1")
plt.ylabel("% de Error")
plt.xlabel("Ã‰poca")
plt.plot(n_epoch, trainingErrorValues_Fold1, label="Training Error")
plt.plot(n_epoch, testErrorValues_Fold1, label="Test Error")
plt.legend()
```

```python
plt.title("Accuracy: FOLD 1")
plt.ylabel("% de Accuracy")
plt.xlabel("Ã‰poca")
plt.plot(n_epoch, trainingAccuracyValues_Fold1, label="Training Accuracy")
plt.plot(n_epoch, testAccuracyValues_Fold1, label="Test Accuracy")
plt.legend()
```
